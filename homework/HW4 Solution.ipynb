```python
def cashflow_matrix(data, face_val=100, coupon_freq='6ME'):
    # assume the treansuries are ID indexed
    data['maturity date'] = pd.to_datetime(data['maturity date'])
    data['quote date'] = pd.to_datetime(data['quote date'])

    all_dates = pd.date_range(
        start=data['quote date'].min() + relativedelta(months=6),
        end=data['maturity date'].max(),
        freq=coupon_freq
    )

    cash_flow_matrix = pd.DataFrame(0, index=data.index, columns=all_dates)

    for idx, row in data.iterrows():
        payment_dates = pd.date_range(
            start=row['quote date'] + relativedelta(months=6),
            end=row['maturity date'],
            freq=coupon_freq
        )

        cpn_payment = row['cpn rate'] / 100 * face_val / 2
        cash_flow_matrix.loc[idx, payment_dates] = cpn_payment

        cash_flow_matrix.loc[idx, row['maturity date']] = face_val + cpn_payment

    return cash_flow_matrix
```

```python
cf_mat = cashflow_matrix(quotes)
```

```python
prices = quotes[['price']].values

cf_mat_inv = np.linalg.inv(cf_mat)

discount_factors = cf_mat_inv @ prices
discount_factors = pd.DataFrame(discount_factors, index=cf_mat.columns, columns=['Discount Factor'])

df_ttm = quotes.set_index('maturity date')[['ttm']]
df_ttm = df_ttm.sort_index()

df131 = pd.merge(df_ttm[['ttm']], discount_factors, left_index=True, right_index=True)

df131['intervals'] = range(1, len(df131) + 1)

df131['continuous discount rate'] = - 2 * np.log(df131['Discount Factor']) / (df131['intervals'])
df131['semiannual discount rate'] = 2 * (1 - np.power(df131['Discount Factor'], 1/df131['intervals']))

df131['ttm'] = df131['intervals'] / 2
```

# 2.1 Solution

```python
plt.figure(figsize=(8, 6))
plt.plot(df131['ttm'], df131['Discount Factor'], marker='o', linestyle='-', label='Discount Factors')
plt.title('Time to Maturity vs. Discount Factors')
plt.xlabel('Time to Maturity (ttm)')
plt.ylabel('Discount Factors')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df131['ttm'], df131['semiannual discount rate'], marker='o', linestyle='-', label='Discount Factors')
plt.title('Time to Maturity vs. Semiannual Discount Factors')
plt.xlabel('Time to Maturity (ttm)')
plt.ylabel('Discount Factors')
plt.grid(True)
plt.legend()
plt.show()
```

# 2.2 Solution

```python
df_for = df131.copy()

df_for['next df'] = df_for['Discount Factor'].shift(-1)
df_for['Forward Discount Factor'] = df_for['next df'] / df_for['Discount Factor']

df_for = df_for.dropna()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df_for['ttm'], df_for['Forward Discount Factor'], marker='o', linestyle='-', label='Discount Factors')
plt.title('Time to Maturity vs. Forward Discount Factors')
plt.xlabel('Time to Maturity (ttm)')
plt.ylabel('Discount Factors')
plt.grid(True)
plt.legend()
plt.show()
```

# 2.3 Solution

```python
df_for['Semiannual Forward Rate'] = 1 - np.power(df_for['Forward Discount Factor'], 2)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df_for['ttm'], df_for['Semiannual Forward Rate'], marker='o', linestyle='-', label='Discount Factors')
plt.title('Time to Maturity vs. Semiannual Forward Discount Factors')
plt.xlabel('Time to Maturity (ttm)')
plt.ylabel('Discount Factors')
plt.grid(True)
plt.legend()
plt.show()
```

# 3. Forward Discount Factors, Forward Rates, and Evolving Spot Rates

Use the data set `famabliss_strips_2024-12-31.xlsx`.

It gives prices on **zero coupon bonds** with maturities of 1 through 5 years.
* These are prices per $1 face value on bonds that only pay principal.
* Such bonds can be created from treasuries by stripping out their coupons.
* In essence, you can consider these prices as the discount factors $Z$, for maturity intervals 1 through 5 years.

In this problem, we focus on six dates: the month of `December` in `2020` through `2024`.

```python
fb_dir = "../data/famabliss_strips_2024-12-31.xlsx"
fb_strips = pd.read_excel(fb_dir).set_index('date')
fb_strips.index = pd.to_datetime(fb_strips.index)
```

### 3.1

For each of the dates, (`Dec 2020`, ..., `Dec 2024`), calculate the forward discount factor, $F(t,T_1,T_2)$, where $T_2 = T_1 + 1$ for $1\le T_1\le 4$.

Make a plot of the curve $F$ at each of these six dates.

```python
for i in range(4):
    name_val = f'Forward {i + 2}-{i + 1}'
    fb_strips[name_val] = fb_strips[i + 2] / fb_strips[i + 1]
```

```python
years = [1, 2, 3, 4]
start_date = datetime(2020, 12, 31)
date_list = [datetime(2020, 12, 31), datetime(2021, 12, 31), datetime(2022, 12, 30), datetime(2023, 12, 29), datetime(2024, 12, 31)]
```

```python
for date in date_list:
    date_row = fb_strips.loc[date][['Forward 2-1', 'Forward 3-2', 'Forward 4-3', 'Forward 5-4']].tolist()

    plt.figure(figsize=(8, 6))
    plt.plot(years, date_row, marker='o', linestyle='-', label='Discount Factors')
    plt.title(f'Years in future vs Forward Discount Factor for {date}')
    plt.xlabel('Years')
    plt.ylabel('Forward Discount Factor')
    plt.grid(True)
    plt.legend()
    plt.show()
```

### 3.2



Use $F$ above to calculate the continuously compounded forward rates, $f$, on each date.

Make a plot of the curve, $f$, at each of these dates.

```python
for i in range(4):
    name1_val = f'Forward {i + 2}-{i + 1}'
    name2_val = f'Forward Rate {i + 2}-{i + 1}'
    fb_strips[name2_val] = 1 - fb_strips[name1_val]
```

```python
for date in date_list:
    date_row = fb_strips.loc[date][['Forward Rate 2-1', 'Forward Rate 3-2', 'Forward Rate 4-3', 'Forward Rate 5-4']].tolist()

    plt.figure(figsize=(8, 6))
    plt.plot(years, date_row, marker='o', linestyle='-', label='Discount Factors')
    plt.title(f'Years in future vs Forward Rate for {date}')
    plt.xlabel('Years')
    plt.ylabel('Forward Rate')
    plt.grid(True)
    plt.legend()
    plt.show()
```

### 3.3

Calculate the continuously compounded spot rates, $r$, at these dates.

Plot both the forward curve and the spot curve on the date of `Dec 2020`.

Recall that $f(t,t,t+1) = r(t,t+1)$.

```python
start_date = datetime(2020, 12, 31)
for_c = fb_strips.loc[start_date][['Forward Rate 2-1', 'Forward Rate 3-2', 'Forward Rate 4-3', 'Forward Rate 5-4']].tolist()
spot_c = (1 - fb_strips.loc[date_list][1]).tolist()

first_rate = 1 - fb_strips.loc[start_date][1]

for_c = [first_rate] + for_c

years = [1, 2, 3, 4, 5]
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(years, for_c, marker='o', color='blue', label='Forward Rate')
ax1.set_xlabel('Years')
ax1.set_ylabel('Rates', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

plt.title('2020 One-year Forward Rates Rates')
fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(years, spot_c, marker='o', color='red', label='Spot Rate')
ax1.set_xlabel('Years')
ax1.set_ylabel('Rates', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

plt.title('2020 One-year Spot Rates')
fig.tight_layout()
plt.show()
```

### 3.4

Plot the forward curve on `Dec 2020` against the time-series evolution of $r(t,t+1)$ for dates $t$ between `Dec 2020` and `Dec 2024`.

How does the evolution of one-year rates between `2020-2024` compare to the forward rates of `2020`?

```python
fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(years, for_c, marker='o', color='blue', label='Forward Rate')
ax1.plot(years, spot_c, marker='o', color='red', label='Spot Rate')
ax1.set_xlabel('Years')
ax1.set_ylabel('Rates', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

plt.title('2020 One-year Forward Rates vs Spot Rates')
fig.tight_layout()
plt.show()
```

# 3.4 Solution

The one year spot rate has changed drastically compared to the expectations of forward rates. Given the recession going on, at the end of 2020 we had historically low Fed Funds rates which were expected to persist. However, Powell aggressively raised rates to curb inflation. This can explain the drastic movement in spot.
